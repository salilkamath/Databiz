```
data = {"Bankrupt" : ["Not Bankrupt", "Bankrupt"]}
value_list = ["Bankrupt"]
for value in df.columns.values[1::]:
  bank_average = bankrupt_df[value].sum()/len(bankrupt_df)
  not_bank_average = not_bankrupt_df[value].sum()/len(not_bankrupt_df)
  data[value] = [not_bank_average, bank_average]
  value_list.append(value)

average_df = pd.DataFrame (data, columns = value_list)

average_df
```
